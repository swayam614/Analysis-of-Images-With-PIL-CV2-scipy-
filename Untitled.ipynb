### Compactness

```python
import cv2
import numpy as np

def calculate_compactness(image_path):
    
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    
    _, thresh = cv2.threshold(image, 200, 255, cv2.THRESH_BINARY)
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    
    if len(contours) > 0:
        contour = max(contours, key=cv2.contourArea)  
        area = cv2.contourArea(contour)
        perimeter = cv2.arcLength(contour, closed=True)

        # compactness =  area/perimeter**2
        compactness = (perimeter **2)/(4 * np.pi * area)
        return compactness
    else:
        return None

# image_path = "/Users/swayampalrecha/Desktop/DIP PROJECT/Gujarati/1.jpg"
image_path = "/Users/swayampalrecha/Desktop/DIP PROJECT/240_F_635410224_GiTFgJxYYtTMk8ai6qZT2hAsotOPAT4d.jpg"
compactness_value = calculate_compactness(image_path)
print("Compactness:", compactness_value)


# The ideal compactness value is often close to 1.0. This indicates that the perimeter of the shape is approximately equal to the perimeter of a circle with the same area, meaning the shape is relatively compact.
```

### Solidity

```python
import cv2

def calculate_solidity(image_path):
    
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    
    _, thresh = cv2.threshold(image, 200, 255, cv2.THRESH_BINARY)

 
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

   
    if len(contours) > 0:
        contour = max(contours, key=cv2.contourArea)  
        area_component = cv2.contourArea(contour)

        
        convex_hull = cv2.convexHull(contour)

        
        area_convex_hull = cv2.contourArea(convex_hull)

        
        solidity = area_component / area_convex_hull
        return solidity
    else:
        return None

# Example usage
image_path = "/Users/swayampalrecha/Desktop/DIP PROJECT/Gujarati/2.jpg"
solidity_value = calculate_solidity(image_path)
print("Solidity:", solidity_value)
```
